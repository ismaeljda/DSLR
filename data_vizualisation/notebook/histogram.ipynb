```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sys
import math
```

```python
data_train = pd.read_csv('../datasets/dataset_train.csv')
houses = data_train['Hogwarts House'].unique()
features_numeriques = data_train.select_dtypes(include=['int64', 'float64']).drop('Index', axis = 1)
features = features_numeriques.columns
features_numeriques['houses'] = data_train['Hogwarts House']
features_numeriques.head()
```

```python
nb_line = math.ceil(len(features) / 5)
fig, axs = plt.subplots(nb_line, 5, figsize=(15, 3 * nb_line))
axs = axs.flatten()

for i, feature in enumerate(features):
    if i < len(axs):
        for house in houses:
            subset = features_numeriques[features_numeriques['houses'] == house][feature].dropna()
            axs[i].hist(subset, bins=30, alpha=0.5, label=house, edgecolor='black')
        
        axs[i].set_title(feature)
        axs[i].legend()

plt.tight_layout()
plt.show()
```
